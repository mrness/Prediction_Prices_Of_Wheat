```python
import plotly.express as ply
import plotly.graph_objects as go
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("Data/KHC.csv", index_col='Date')
data2 = pd.read_csv("Data/KHC.csv", index_col="High")
```

```python
data = data.dropna().drop_duplicates()
data2.sort_index().drop("Open","Low","Close","Adj Close", "Volume")
```

```python
data.drop(["Open","High","Low","Adj Close","Volume"], axis=1, inplace=True)
```

```python
data2.drop(["H"])
```

```python
KHC_line = ply.line(data,x=data.index,y='Close')
KHC_line.show()
```
